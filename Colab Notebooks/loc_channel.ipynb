```python
import jsonlines, json
```

```python
def read_jsonl_file(path):
    data = []
    
    for line in open(path, 'r', encoding='utf8'):
        # print(1)
        try:
            data.append(json.loads(line))
        except Exception as e:
            # print(e)
            pass
    return data
```

```python
data = read_jsonl_file('final_channel_urls.jsonl')
```
